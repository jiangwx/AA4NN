```python
import numpy as np
from numpy import array
from numpy.random import normal
import matplotlib.pyplot as plt
import os, sys
import struct
```

解析AA的输出

```python
aa_output_major = np.zeros((1000,256),dtype=np.float32)#1000是输出的个数，256是epsilon的个数
f = open("../AA_Jaffe/cmake-build-debug/fc8.txt","r")   #设置文件对象
lines = f.readlines()#读取全部内容
lines_cnt = len(lines)#每个txt文件的行数
line_length = len(lines[0].split('+'))#每一行的元素数量
for i in range(lines_cnt):
    temp = lines[i].split('+')#根据+位置将每一行数据进行分割
    for j in range(line_length):
        if (j == 0):
            value = float(temp[j])#每一行的第一个数据
        elif (j ==line_length - 1):
            value = float(temp[j][:temp[j].rfind('e')])
            index = int(temp[j][temp[j].rfind('e')+1:temp[j].rfind('[')])
            aa_output_major[i,index-1] = value
        else:
            #以-7.65585e-05e3为例，-7.65585e-05是值，e3是epsilon3的缩写
            value = float(temp[j][:temp[j].rfind('e')])#取最后一个e前面的字符转换成float型
            index = int(temp[j][temp[j].rfind('e')+1:])#取最后一个e后面的字符转换成int型作为epsilon的下标，有的epsilon被砍掉了
            aa_output_major[i,index-1] = value
print aa_output_major
```

```python
aa_output_abs = np.absolute(aa_output_major)
epsilon_average = np.mean(aa_output_abs, axis=0)
print epsilon_average
```
